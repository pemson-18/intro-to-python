## Section 1: Python as a Scientific Calculator (10 minutes)

### Theory
Python can perform mathematical operations directly. It supports:
- Basic arithmetic: `+`, `-`, `*`, `/`
- Integer division: `//` (quotient only)
- Modulo: `%` (remainder)
- Exponentiation: `**`
- Mathematical functions via `math` module

### Example

### Import Statements (Using External Code)

Importing lets you reuse optimized, well-tested code from Python's standard library or external packages instead of reinventing it.

Common forms:
- `import math` → access with `math.pi`, `math.sqrt(9)` (keeps namespace clean).
- `from math import sqrt, pi` → direct names `sqrt(9)`, `pi` (fewer keystrokes, higher collision risk).
- `import math as m` → alias for brevity: `m.sqrt(9)`.
- `import numpy as np` → conventional alias for NumPy.

Guidelines:
- Prefer `import module` for clarity in teaching (`math.sqrt` shows the source).
- Use `from ... import ...` sparingly for very frequently used symbols.
- Avoid `from module import *` (pollutes namespace, obscures origins).
- Group standard library imports first, then third-party, then local modules.

Behind the scenes, Python loads a module once and caches it in `sys.modules`; repeated imports are fast. If a module is not found you will see `ModuleNotFoundError`—install or correct the name.

Quick example:
```python
import math
radius = 5
area = math.pi * radius**2
print(area)
```


````python
# Example: Scientific calculations
import math

# Calculate the area of a circle with radius 5
radius = 5
area = math.pi * radius**2
print(f"Area of circle: {area:.2f} square units")

# Calculate force (F = ma)
mass = 10  # kg
acceleration = 9.8  # m/s²
force = mass * acceleration
print(f"Force: {force} N")
````

## Section 2: Data Structures for Science (10 minutes)

### Theory: Lists - The Foundation of Data Collections
Lists are ordered, mutable collections that can store scientific data:
- **Creating lists:** `data = [1, 2, 3, 4, 5]`
- **Indexing:** `data[0]` (first element), `data[-1]` (last element)
- **Slicing:** `data[1:3]` (elements at index 1 and 2)

## Section 2B: Dictionaries - Key-Value Data Structures (8 minutes)

### Theory: Organizing Related Data
Dictionaries store data as key-value pairs, perfect for organizing related scientific information:
- **Creating dictionaries:** `data = {'key1': value1, 'key2': value2}`
- **Accessing values:** `data['key1']`
- **Adding/updating:** `data['new_key'] = new_value`
- **Methods:** `keys()`, `values()`, `items()`, `get()`

**Why use dictionaries in science?**
- Store related measurements (temperature, pressure, volume)
- Organize experimental parameters
- Map element symbols to properties
- Store student records or sample data

### Example

````python
# Example: Chemical element properties
element = {
    'name': 'Oxygen',
    'symbol': 'O',
    'atomic_number': 8,
    'atomic_mass': 15.999,
    'state': 'gas'
}

print(f"Element: {element['name']} ({element['symbol']})")
print(f"Atomic Number: {element['atomic_number']}")
print(f"Atomic Mass: {element['atomic_mass']} u")

# Add new property
element['electronegativity'] = 3.44
print(f"\nAll properties: {element}")

# Example: Experimental data
experiment = {
    'temperature': 25.0,  # Celsius
    'pressure': 101.3,    # kPa
    'volume': 2.5,        # Liters
    'pH': 7.0
}

print(f"\nExperimental Conditions:")
for key, value in experiment.items():
    print(f"  {key.capitalize()}: {value}")
````

## Section 3: Control Flow - Loops and Conditionals (10 minutes)

### Theory: Making Decisions and Repeating Tasks
**Conditionals** allow programs to make decisions:
```python
if condition:
    # execute if True
elif another_condition:
    # execute if first is False and this is True
else:
    # execute if all above are False
```

**Loops** repeat operations:
- `for` loop: Iterate over sequences
- `while` loop: Repeat while condition is True

### Example

````python
# Example: Classify exam scores
scores = [85, 92, 67, 78, 95, 54, 88]

for score in scores:
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    print(f"Score {score} → Grade {grade}")
````

````python
# Example: using range() in a for loop
# range(start, stop, step)
for n in range(0, 11, 2):  # 0 to 10 inclusive, step 2
  print(f"Even number: {n}")

# Summing a sequence with range
total = 0
for i in range(1, 6):  # 1..5
  total += i
print(f"Sum 1..5 = {total}")
````

## Section 4: Functions - Reusable Scientific Tools (10 minutes)

### Theory: Creating Your Own Tools
Functions encapsulate reusable code:
```python
def function_name(parameters):
    # code block
    return result
```

**Benefits for Scientific Computing:**
- Reusability (write once, use many times)
- Modularity (break complex problems into smaller parts)
- Testing (test individual components)
- Documentation (clear purpose and usage)

### Example

````python
# Example: Convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    """Convert temperature from Celsius to Fahrenheit."""
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Test the function
temp_c = [0, 20, 37, 100]
for temp in temp_c:
    temp_f = celsius_to_fahrenheit(temp)
    print(f"{temp}°C = {temp_f}°F")
````

### Using `range()` Inside Functions

The built-in `range()` is commonly used in functions that perform repeated calculations.

Syntax patterns:
- `range(stop)` → 0 .. stop-1
- `range(start, stop)` → start .. stop-1
- `range(start, stop, step)` → step increments (can be negative)

Key points:
- Returns a lazy sequence (not a list). Convert with `list(range(...))` if you need all values at once.
- Efficient for large loops since it does not build a full list.
- Use negative steps for reverse iteration: `range(10, 0, -1)`.
- Pair with `len()` for index-based loops: `for i in range(len(data)):`.
- Prefer direct iteration (`for item in data:`) unless you need the index.

Common patterns:
```python
# Summation
s = 0
for i in range(1, 6):  # 1..5
    s += i

```

Below: A function using `range()` to approximate π via Leibniz series.

````python
# Example: using range() in a for loop
# range(start, stop, step)
for n in range(0, 11, 2):  # 0 to 10 inclusive, step 2
  print(f"Even number: {n}")

# Summing a sequence with range
total = 0
for i in range(1, 6):  # 1..5
  total += i
print(f"Sum 1..5 = {total}")
````
